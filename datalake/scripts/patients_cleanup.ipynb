```python
import pandas as pd
import datetime
from dateutil.parser import parse
```

```python
def convert_date(date_string):
    formats = ['%d/%m/%y', '%m/%d/%y', '%m.%d.%y', '%d.%m.%Y', '%m/%d/%Y', '%m-%d-%Y']
    for fmt in formats:
        try:
            date_obj = datetime.datetime.strptime(str(date_string), fmt)
            return date_obj.strftime('%Y-%m-%d')
        except ValueError:
            pass
    return None
```

```python
def conver_fuzzy_date(date_string):
    date_string_s = str(date_string)
    return parse(date_string_s, fuzzy=True)
```

```python
df = pd.read_csv('./patients_final.csv')
```

```python
df['BIRTHDATE'] = df['BIRTHDATE'].apply(conver_fuzzy_date)
```

```python
df['DEATHDATE'] = df['DEATHDATE'].apply(convert_date)
```

```python
# print(df[['DEATHDATE']].to_string(index=False))
```

```python
# Cleanup newline characters within columns
df = df.replace('\n','', regex=True)
```

```python
df.to_csv('./patients_final_2.csv',doublequote=True)
```
